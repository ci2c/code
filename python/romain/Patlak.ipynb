```python
aif=np.loadtxt('/NAS/tupac/romain/git/ROCKETSHIP/DUBUIS_MYRIAM/processing1/aif.txt')
img=nib.load("/NAS/tupac/romain/git/ROCKETSHIP/DUBUIS_MYRIAM/processing1/C_t_wonan.nii.gz")
imgT10=nib.load("/NAS/tupac/romain/git/ROCKETSHIP/DUBUIS_MYRIAM/processing1/T1_0.nii.gz")

img_data=img.get_data()

aif_lin=aif[8:]
img_data_lin=img_data[:,:,:,8:]

X=np.cumsum(aif_lin)/aif_lin
Xs=X.reshape(40,1)
Y=img_data_lin/aif_lin

Slope=np.empty_like(imgT10)
Intercept=np.empty_like(imgT10)
Error=np.empty_like(imgT10)
for i in range(Y.shape[0]):
    for j in range(Y.shape[1]):
        for k in range(Y.shape[2]):            
            reg = LinearRegression().fit(Xs,Y[i,j,k,:])            
            Error[i,j,k]=reg.score(Xs,Y[i,j,k,:])
            Slope[i,j,k]=reg.coef_
            Intercept[i,j,k]=reg.intercept_
            
array_img = nib.Nifti1Image(Slope, imgT10.affine)
nib.save(array_img,"/NAS/tupac/romain/git/ROCKETSHIP/DUBUIS_MYRIAM/processing1/Ktrans_slope.nii.gz")
array_img = nib.Nifti1Image(Intercept, imgT10.affine)
nib.save(array_img,"/NAS/tupac/romain/git/ROCKETSHIP/DUBUIS_MYRIAM/processing1/Vp_intercept.nii.gz")
array_img = nib.Nifti1Image(Error, imgT10.affine)
nib.save(array_img,"/NAS/tupac/romain/git/ROCKETSHIP/DUBUIS_MYRIAM/processing1/error.nii.gz")        
```

```python
import os, sys, re
from sklearn.linear_model import LinearRegression
import nibabel as nib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

DIR="/NAS/dumbo/protocoles/CAR-T-Cell/process/DUBUIS_MYRIAM/"#sys.argv[1]
print(DIR)
Hct=0.45
C_b=np.loadtxt(DIR+"aif.txt")
print(C_b)
C_p=C_b / (1 - 0.45)
print(C_p)
C_p=C_b / (1 - Hct)
print(C_p)
```

```python
img=nib.load(DIR+"C_t_wonan.nii.gz")
imgT10=nib.load(DIR+"template.nii.gz")

img_data=img.get_data()

#Recupre uniquement les dynamiques au-dessus de la 8ime por n'avoir que la partie "wash-out"
aif_lin=aif[8:]
img_data_lin=img_data[:,:,:,8:]

X=np.cumsum(aif_lin)/aif_lin
Xs=X.reshape(len(aif_lin),1)
Y=img_data_lin/aif_lin

Slope=np.empty_like(imgT10.get_data())
Intercept=np.empty_like(imgT10.get_data())
Error=np.empty_like(imgT10.get_data())

for i in range(Y.shape[0]):
    for j in range(Y.shape[1]):
        for k in range(Y.shape[2]):
            reg = LinearRegression().fit(Xs,Y[i,j,k,:])
            Error[i,j,k]=reg.score(Xs,Y[i,j,k,:])
            Slope[i,j,k]=reg.coef_
            Intercept[i,j,k]=reg.intercept_s

array_img = nib.Nifti1Image(Slope, imgT10.affine)
nib.save(array_img,DIR+"Ktrans_slope.nii.gz")
array_img = nib.Nifti1Image(Intercept, imgT10.affine)
nib.save(array_img,DIR+"Vp_intercept.nii.gz")
array_img = nib.Nifti1Image(Error, imgT10.affine)
nib.save(array_img,DIR+"error.nii.gz")
```
