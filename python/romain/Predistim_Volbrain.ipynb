```python
def getAgeAndSexe(ptID):    
    fh = open(os.path.join(STUDY_PATH,'DemographicData20200406.csv'))
    reader = csv.reader(fh, delimiter = ';')

    for ligne in reader:
        if ligne[0]==ptID:
            break

    return ligne[1], ligne[2]
#df = pd.read_csv(file_name,sep = ';')
```

```python
def getCQT1AndMultiTE(ptID):
    fh = open(os.path.join(STUDY_PATH,'CQData20200406.csv'))
    reader = csv.reader(fh, delimiter = ';')

    for ligne in reader:
        if ligne[0]==ptID:
            break

    return ligne[1], ligne[2]
#df = pd.read_csv(file_name,sep = ';')
```
