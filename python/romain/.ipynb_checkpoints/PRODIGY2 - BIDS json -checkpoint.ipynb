```python
#Ajout des SliceTiming
import os
import json
import numpy as np

#45 coupes interleaved cf. page ci-dessus
tmp_tab_order= np.array([1,8,15,22,29,36,43,2,9,16,23,30,37,44,3,10,17,24,31,38,45,4,11,18,25,32,39,5,12,19,26,33,40,6,13,20,27,34,41,7,14,21,28,35,42])
TA=(2.4/45)
values=tmp_tab_order*TA-TA

for root, dirs, files in os.walk("/NAS/tupac/protocoles/prodigy2/BIDS_DB/sub-DELAMBREARMELLEPRODIGY20180927/"):    
    for file in files:
        if file.endswith('bold.json') & file.startswith('sub-') :
            filename=os.path.join(root,file)
            with open(filename,'r') as f:
                json_data = f.read()
                f.close()
            data = json.loads(json_data)
            print(filename)
            if not data.has_key('SliceTiming') :
                print('SliceTiming')
                data['SliceTiming']=list(values)
                with open(filename, 'w') as outfile:
                    json.dump(data, outfile,sort_keys=True,indent=4,separators=(',', ': '))
            else :
                print(data['SliceTiming'])
```

```python
#Ajout de PhaseEncodingDirection / TotalReadoutTime / IntendedFor dans les fichiers json de fmap
#https://neurostars.org/t/questions-to-the-output-of-fmriprep-regarding-slice-timing-correction-and-distortion-correction/1628/3
import os
import json
import numpy as np
import glob 

for root, dirs, files in os.walk("/NAS/tupac/protocoles/prodigy2/BIDS_DB/sub-TERRIERPATRICKPRODIGYTEMOIN20181122/"):    
    for file in files:
        if file.endswith('epi.json') & file.startswith('sub-') :
                filename=os.path.join(root,file)
                with open(filename,'r') as f:
                    json_data = f.read()
                    f.close()
                data = json.loads(json_data)
                #PhaseEncodingDirection
                #Attention je ne sais pas le sens entre j et j- "PhaseEncodingDirection": "j-",                              
                #https://neurostars.org/t/determining-bids-phaseencodingdirection-from-dicom/612/8
                #norme utilisé A vers P sens +0- donc j- et de P vers A sens -0+ donc j
                if filename.find("_dir-PA") > 0 :
                    data['PhaseEncodingDirection']="j"
                elif filename.find("_dir-AP") > 0 :
                    data['PhaseEncodingDirection']="j-"
                else :
                    print("sushi")
                
                #TotalReadoutTime                
                if data.has_key('PixelBandwidth') & data.has_key('ReconMatrixPE') :
                    data['TotalReadoutTime']=0.02137748033808136500 
                    print(1./(data['PixelBandwidth']/data['ReconMatrixPE'])) #avant je le calculer comme cela
                else :
                    print("sushi")              
                
                #"IntendedFor": "func/sub-01_task-motor_bold.nii.gz"                
                files_tmp = glob.glob(root.replace("fmap","func")+'/*_bold.nii.gz') #ou text_files = [f for f in os.listdir(path) if f.endswith('.txt')]
                files_list=[]
                for fil in files_tmp:
                    files_list.append('func/'+os.path.basename(fil))              
                print(files_list)
                data['IntendedFor']=files_list                
                with open(filename, 'w') as outfile:
                    json.dump(data, outfile,sort_keys=True,indent=4,separators=(',', ': '))     
```

```python
#Ajout des PhaseEncodingDirection pour les func/
#et Effective Echo Spacing https://github.com/rordenlab/dcm2niix/issues/130 
#https://support.brainvoyager.com/brainvoyager/functional-analysis-preparation/29-pre-processing/78-epi-distortion-correction-echo-spacing-and-bandwidth
#https://www.jiscmail.ac.uk/cgi-bin/webadmin?A2=fsl;a66ccadf.1404
#EPI factor equivalent à "EchoTrainLength": 31,

import os
import json
import numpy as np

for root, dirs, files in os.walk("/NAS/tupac/protocoles/prodigy2/BIDS_DB/sub-DUCLOSJEANprodigy20180626/"):    
    for file in files:
        if file.endswith('bold.json') & file.startswith('sub-') :
            filename=os.path.join(root,file)
            with open(filename,'r') as f:
                json_data = f.read()
                f.close()
            data = json.loads(json_data)
            if not data.has_key('PhaseEncodingDirection') :
                data['PhaseEncodingDirection']="j"    
            if not data.has_key('EffectiveEchoSpacing') :
                data['EffectiveEchoSpacing']=0.0007230996698207571                  
            with open(filename, 'w') as outfile:
                json.dump(data, outfile,sort_keys=True,indent=4,separators=(',', ': '))          

        if file.endswith('epi.json') & file.startswith('sub-') :            
            filename=os.path.join(root,file)
            with open(filename,'r') as f:
                json_data = f.read()
                f.close()
            data = json.loads(json_data) 
            if not data.has_key('TotalReadoutTime') :
                data['TotalReadoutTime']=0.02137748033808136500
            with open(filename, 'w') as outfile:
                json.dump(data, outfile,sort_keys=True,indent=4,separators=(',', ': '))
```

```python
import os
import json
import numpy as np
import glob 
#FMRI_CompleteJsonFile
#dcmdump +P 2001,1018 +P 0018,0080 +P 0018,9069 +P 0018,0087 +P 2001,1013 +P 2001,1022 '/NAS/tupac/protocoles/prodigy2/data/TERRIER^PATRICK_PRODIGY_TEMOIN_2018-11-22/1201_SDMT_FMRI/MR_1.3.46.670589.11.34037.5.0.4380.2018112210022888648.dcm'

Nslices=45
WFS=9.9012508
ETL=31
FS=3
PRFIP=2
TRsec=2.4

sliceorder = np.array([]);
space = round(np.sqrt(Nslices));
for k in np.arange(1,space+1):
    tmp=np.arange(k,Nslices+1,space)
    sliceorder=np.concatenate([sliceorder,tmp])
TA = TRsec / Nslices
slicetiming = np.arange(0,2.4,TA)
stc=[]
for k in np.arange(1,Nslices+1):
    stc.append(round(slicetiming[np.where(sliceorder==k)][0],10))

for root, dirs, files in os.walk("/NAS/tupac/protocoles/prodigy2/BIDS_DB/sub-DUCLOSJEANprodigy20180626"):    
    for file in files:
        if file.endswith('bold.json') & file.startswith('sub-') :
            filename=os.path.join(root,file)
            with open(filename,'r') as f:
                json_data = f.read()
                f.close()
            data = json.loads(json_data)  
            #data['EffectiveEchoSpacing']=0.00071258267793604550 
            data['SliceTiming']=list(stc)
            with open(filename, 'w') as outfile:
                json.dump(data, outfile,sort_keys=True,indent=4,separators=(',', ': '))
```
