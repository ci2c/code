Imports

```python
import nibabel as nib
import numpy as np
from dipy.segment.bundles import RecoBundles
from dipy.viz import actor, window
from dipy.viz.colormap import line_colors
from dipy.io.dpy import Dpy
from dipy.tracking.streamline import transform_streamlines
from dipy.tracking.streamline import select_random_set_of_streamlines
from dipy.io.image import load_nifti, save_nifti
from dipy.io import read_bvals_bvecs
from dipy.core.gradients import gradient_table
from dipy.reconst.dti import TensorModel
from dipy.viz import colormap
from dipy.reconst.dti import fractional_anisotropy, color_fa, lower_triangular
from dipy.segment.metric import ResampleFeature
from dipy.segment.metric import AveragePointwiseEuclideanMetric
from dipy.segment.metric import CenterOfMassFeature
from dipy.segment.metric import EuclideanMetric
from dipy.segment.metric import IdentityFeature
from dipy.segment.clustering import QuickBundles
```

Variables globales

```python
NAS="tupac"
protocole="alexcis"
DIR_PATH="/NAS/"+NAS+"/protocoles/"+protocole
FS_PATH=DIR_PATH+"/FS53/"
bundle = nib.streamlines.load(FS_PATH+"p_AR28_na/dti/whole_brain_6_1500000.trk")
```

Ecriture du fichier dans un format propriétaire à dipy permettant une lecture partiel

```python
dpw = Dpy(FS_PATH+"p_AR28_na/dti/whole_brain_6_1500000.dpy", 'w')
dpw.write_tracks(bundle.streamlines)
dpw.close()
```

Affichage d'un extrait de la tracto cerveau entier et du corps calleux

```python
plot_tracto = select_random_set_of_streamlines(bundle.streamlines,9999)
streamlines_actor = actor.line(plot_tracto, line_colors(plot_tracto))

ren = window.ren()
ren.add(streamlines_actor)
data, affine = load_nifti(FS_PATH+"p_AR28_na/dti/CorpsCalleux_mask.nii.gz")
surface_opacity = 0.5
surface_color = [0, 1, 1]
seedroi_actor = actor.contour_from_roi(data, affine,surface_color, surface_opacity)
ren.add(seedroi_actor)
window.show(ren)
```

Calcul de la FA et lecture de la FA précedemment calculée

```python
data, affine = load_nifti(FS_PATH+"p_AR28_na/dti/dti.nii.gz")
bvals, bvecs = read_bvals_bvecs(FS_PATH+"p_AR28_na/dti/dti.bval",FS_PATH+"p_AR28_na/dti/dti.bvec")
gtab = gradient_table(bvals, bvecs)
tenmodel = TensorModel(gtab)
tenfit = tenmodel.fit(data)

nii = nib.load(FS_PATH+"p_AR28_na/dti/dti_finalcor_FA.nii.gz")
```
