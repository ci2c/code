```python
mask10 = load('/NAS/dumbo/protocoles/CogPhenoPark/CortThick_Quentin/PALM/Surfacique/G2vsG4/ocn_rh_10_G2vsG4.mgh')
mask15 = load('/NAS/dumbo/protocoles/CogPhenoPark/CortThick_Quentin/PALM/Surfacique/G2vsG4/ocn_rh_10_G2vsG4.mgh')
mask20 = load('/NAS/dumbo/protocoles/CogPhenoPark/CortThick_Quentin/PALM/Surfacique/G2vsG4/ocn_rh_10_G2vsG4.mgh')
mask10_data=mask10.get_data()
mask15_data=mask15.get_data()
mask20_data=mask20.get_data()
for root, dirs, files in os.walk("//NAS/dumbo/protocoles/CogPhenoPark/CortThick_Quentin/PALM/data/"):    
    for file in files:
        if file.endswith('_rh.thickness.fwhm10.fsaverage.mgh'): 
            pos=file.find('_rh.thic')
            namePat=file[4:pos]
            resu=namePat
            data=load(root+file)
            data_data=data.get_data()
            resu=resu+" "+str(np.mean(data_data[mask10_data==1]))
            
            data_data=data.get_data()
            resu=resu+" "+str(np.mean(data_data[mask10_data==2]))
            
            data=load(root+"sub-"+namePat+"_rh.thickness.fwhm15.fsaverage.mgh")            
            data=load(root+file)
            data_data=data.get_data()
            resu=resu+" "+str(np.mean(data_data[mask15_data>0]))
            
            data=load(root+"sub-"+namePat+"_rh.thickness.fwhm20.fsaverage.mgh")
            data_data=data.get_data()
            resu=resu+" "+str(np.mean(data_data[mask20_data>0]))
            print(resu)
```

```python
#for file in `ls G*.nii`; do mri_convert -rt nearest --upsample 2 $file u_$file; done
for root, dirs, files in os.walk("/NAS/dumbo/protocoles/CogPhenoPark/CortThick_Quentin/Shape/Data_Groupes/LeftCaudate/"):    
    for file in files: 
        if file.endswith('.nii') and file.startswith('u_'): 
            #outputCmd = os.popen('mri_convert -rt nearest --upsample 2 '+os.path.join(root,file)+' '+os.path.join(root,"resized_"+file),'r')
            outputCmd = os.popen('fslstats ' + os.path.join(root,file) +' -c','r')
            cog=outputCmd.read()
            cogS=cog.split(" ")
            img = nib.load(os.path.join(root,file))
            img.affine[0][3]=img.affine[0][3]-np.float(cogS[0])
            img.affine[1][3]=img.affine[1][3]-np.float(cogS[1])
            img.affine[2][3]=img.affine[2][3]-np.float(cogS[2])
            nib.save(img,os.path.join(root,"rc_"+file))
            print(os.path.join(root,"r"+file))
    break
```
